```python
import re

for m in range(1):
#     driver.get(lis[m])
    driver.implicitly_wait(3)
    try:
        sess=driver.find_element(By.XPATH,'/html/body/div[2]/div[3]/section/div[2]/header/div/div[3]/div[1]/div/h1').text
    except:
        sess=''
    try:
        sub=driver.find_element(By.CLASS_NAME,'full-page-title').text
    except:
        sub=''
    try:
        h3_element = driver.find_element(By.XPATH,"//h3[text()='Date']")
        date = h3_element.find_element(By.XPATH,"following-sibling::span").text
    except:
        date=''
    try:
        pre = driver.find_element(By.XPATH,"//h3[text()='Presenters']")
        presenter = pre.find_element(By.XPATH,"following-sibling::span").text
        SESSION.append(sess)
        SUB_SESSION.append(sub)
        NAMES.append(presenter)
        ROLE.append('Presenter')
        AFFILIATION.append('')
        DATE.append(date)
    except:
        SESSION.append(sess)
        SUB_SESSION.append(sub)
        NAMES.append('')
        ROLE.append('')
        AFFILIATION.append('')
        DATE.append(date)
    try:
        h3_element = driver.find_element(By.XPATH,"//h3[text()='Authors']")
        authors = h3_element.find_element(By.XPATH,"following-sibling::span")
        a=authors.text
        result=[]
        if(',' not in a):
            result.append(a)
        else:
            matches = re.findall(r'(\D+)(\d+)', a)
            result = [[name.strip(), digit] for name, digit in matches]
        span_element = driver.find_element(By.XPATH,'//span[@data-toggle="collapse" and contains(@class, "collapse-data-control")]')
        
        
        actions = ActionChains(driver)
        actions.move_to_element(span_element).perform()
        span_element.click()
        
        driver.implicitly_wait(2)
        h3_element = driver.find_element(By.XPATH,"//h3[text()='Author affiliations']")
        aff_div = h3_element.find_element(By.XPATH,"following-sibling::div")
        affs=aff_div.find_elements(By.TAG_NAME,'li')
        temp_au=[]
        temp_af=[]
        if(len(result)==1):
            temp_au.append(result[0])
            temp_af.append(affs[0].text)
        else:
            for i in range(len(result)):
                for j in range(len(affs)):
                    text=affs[j].text
                    match = re.search('\d', text)
                    part1 = text[:1].strip()
                    part2 = text[1:].strip()
                    if(result[i][1] in part1):
                        temp_au.append(result[i][0])
                        temp_af.append(part2)
                    
        for i in range(len(temp_au)):
            NAMES.append(temp_au[i])
            ROLE.append('Author')
            AFFILIATION.append(temp_af[i])
            SESSION.append(sess)
            SUB_SESSION.append(sub)
            DATE.append(date)
    except:
        print('error at ',lis[m])
        print(m)
    print(m,'.')
```
