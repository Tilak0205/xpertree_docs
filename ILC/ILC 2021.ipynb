```python
for m in range(len(days)):
    driver.get(days[m])
    driver.implicitly_wait(3)
    chairs=[]
    roles=[]
    f1=0
    for c in range(1,10):
        try:
            cname=driver.find_element(By.XPATH,'/html/body/div[4]/div[5]/div/div/table/tbody/tr['+str(c)+']/td[2]/table/tbody/tr/td[1]/a').text
            chairs.append(cname)
            f1=1
            try:
                role=driver.find_element(By.XPATH,'/html/body/div[4]/div[5]/div/div/table/tbody/tr['+str(c)+']/td[2]/table/tbody/tr/td[2]').text
                roles.append(role)
            except:
                roles.append('Chair')
        except:
            break
    try:
        dm=driver.find_element(By.XPATH,'/html/body/div[4]/div[4]/div/div/span[6]/span').text
    except:
        dm=''
        
    ses=driver.find_element(By.CLASS_NAME,'container-fluid')
    session=ses.find_element(By.TAG_NAME,'h4').text
    for ch in range(len(chairs)):
        SESSION.append(session)
        SUB_SESSION.append('Introduction')
        DATE_TIME.append(dm)
        NAME.append(chairs[ch])
        ROLE.append(roles[ch])
    f2=0
    for s in range(1,20):
        try:
            ss=driver.find_element(By.XPATH,'/html/body/div[4]/div[6]/div/div/div/table/tbody/tr['+str(s)+']/td[3]/span').text
            speakers=[]
            f2=1
            try:
                d=driver.find_element(By.XPATH,'/html/body/div[4]/div[6]/div/div/div/table/tbody/tr['+str(s)+']/td[4]/span').text
            except:
                d=''
            try:
                li=driver.find_element(By.XPATH,'/html/body/div[4]/div[6]/div/div/div/table/tbody/tr['+str(s)+']/td[3]/div')
                total=li.find_elements(By.TAG_NAME,'a')
                for i in range(len(total)):
                    speakers.append(total[i].text)
                for i in range(len(speakers)):
                    SESSION.append(session)
                    SUB_SESSION.append(ss)
                    DATE_TIME.append(d)
                    NAME.append(speakers[i])
                    ROLE.append('Speaker')
                    
            except:
                SESSION.append(session)
                SUB_SESSION.append(ss)
                DATE_TIME.append(d)
                NAME.append('')
                ROLE.append('')
            print(m,s)
                
        except:
            break
    if(f1==0 and f2==0):
        SESSION.append(session)
        SUB_SESSION.append('Introduction')
        DATE_TIME.append(dm)
        NAME.append('')
        ROLE.append('')
    print('p: ',m)
```

```python
import pandas as pd
```

```python
tt = pd.ExcelFile('ILC_2021.xlsx').parse('Sheet1')
sub=tt['SUB_SESSION']
names=tt['NAMES']
```

```python
Initial=[]
First=[]
Middle=[]
Last=[]
```

```python
def NameDivider(Name):
    inti=""
    first=""
    last=""
    midd=[]
    temp=str(Name).split(" ")
    for i in range(len(temp)):
        if(len(temp[i])==2 and i==1):
            if(temp[i][1]=='.'):
                inti=temp[i][0]
            continue
        if(i==0):
           first=temp[i]
           continue
        if(i==len(temp)-1):
           last=temp[i]
           continue
        else:
           midd.append(temp[i])
    if(midd==[]):
           mid=""
    if(midd!=[]):
           mid=" ".join(midd)
    return inti,first,mid,last


for j in range(len(names)):
    inti,first,mid,lst=NameDivider(names[j])
    Initial.append(inti)
    First.append(first)
    Middle.append(mid)
    Last.append(lst)
```

```python
search=['Gadolinium',
'MRI+Contrast Agent',
'Focal liver lesions',
'Liver metastases',
'Nephrogenic Systemic Fibrosis',
'Colorectal Cancer+Liver Metastasis'
'Breast Cancer+Liver Metastasis',
'Lung Cancer+Liver Metastatsis',
'Gastric Cancer+Liver Metastasis',
'Primary Liver Cancer',
'Gadolinium toxicity',
'Gadolinium deposition']
```

```python
s_ele=[]
found=[]
```

```python
for i in range(len(sub)):
    f=-1
    for j in range(len(search)):
        if(('+' not in search[j]) and (type(sub[i])!=float)):
            if(search[j].lower() in sub[i].lower()):
                f=j
        else:
            if(type(sub[i])!=float):
                temp=search[j].split('+')
                if((temp[0] in sub[i].lower()) and (temp[1] in sub[i].lower())):
                    f=j
    if(f!=-1):
        s_ele.append(search[j])
        found.append('Found')
    else:
        s_ele.append('No search term')
        found.append('Not found')
```

```python
bt=pd.DataFrame()
bt['Session ST']=found
bt['InSession']=s_ele
# bt['Abstract ST']=ab_found
# bt['InAbstract']=ab_s_ele
bt['Initial']=Initial
bt['First Name']=First
bt['Middle Name']=Middle
bt['Last Name']=Last
```

```python
bt.to_excel('search_terms_2021.xlsx')
```
