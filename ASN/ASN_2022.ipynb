```python
def NameDivider(Name):
    inti=""
    first=""
    last=""
    midd=[]
    temp=str(Name).split(" ")
    for i in range(len(temp)):
        if(len(temp[i])==2 and i==1):
            if(temp[i][1]=='.'):
                inti=temp[i][0]
            continue
        if(i==0):
           first=temp[i]
           continue
        if(i==len(temp)-1):
           last=temp[i]
           continue
        else:
           midd.append(temp[i])
    if(midd==[]):
           mid=""
    if(midd!=[]):
           mid=" ".join(midd)
    return inti,first,mid,last


for j in range(len(names)):
    inti,first,mid,lst=NameDivider(names[j])
    Initial.append(inti)
    First.append(first)
    Middle.append(mid)
    Last.append(lst)
```
