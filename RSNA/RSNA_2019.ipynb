```python
import requests
from PyPDF2 import PdfFileReader

url = 'https://www.rsna.org/-/media/Files/RSNA/Annual-meeting/Program/RSNA-2020-program.ashx'
response = requests.get(url)

# Open the PDF file
pdf = PdfFileReader(response.content)

# Get the number of pages
num_pages = pdf.getNumPages()

# Iterate through each page
for i in range(num_pages):
    page = pdf.getPage(i)
    text = page.extractText()
    print(text)
```

```python
from PyPDF2 import PdfReader

# Open the PDF file
with open("RSNA-2020-program.pdf", "rb") as file:
    # Create a PdfReader object
    pdf_reader = PdfReader(file)
    
    # Get the number of pages in the PDF
    num_pages = len(pdf_reader.pages)
    print("Number of pages:", num_pages)
    
    # Iterate through each page
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]
        print("Page", page_num + 1, "contents:", page.extract_text())
```
