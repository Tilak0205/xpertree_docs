```python
%%notify

for i in range(len(links)):
# for i in [22,59,60,79,116,121,124,125,144,149,151,198,200,205,209,243,257]:
    driver.get(links[i])
    WebDriverWait(driver,20).until(EC.element_to_be_clickable((By.XPATH,"/html/body/div[2]/div/div[2]/div[1]/ul/li[3]/a/div/div/div[1]/div")))
    ss=driver.find_element(By.CLASS_NAME,"object-detail-title").text
    try:
        rl=[]
        driver.find_element(By.CLASS_NAME,"line-one")
        rl1=driver.find_elements(By.CLASS_NAME,"line-one")
        for s in rl1:
            rl.append(s.text)
    except:
        rl=[]
    try:
        ch=[]
        driver.find_element(By.CLASS_NAME,"line-two")
        ch1=driver.find_elements(By.CLASS_NAME,"line-two")
        for s in ch1:
            ch.append(s.text)
    except:
        ch=[]
    try:
        ca=[]
        driver.find_element(By.CLASS_NAME,"line-three")
        ca1=driver.find_elements(By.CLASS_NAME,"line-three")
        for s in ca1:
            ca.append(s.text)
    except:
        ca=[]
    try:
        li=driver.find_element(By.CLASS_NAME,"object-detail-inline-bookmark-link")
        li=driver.find_elements(By.CLASS_NAME,"object-detail-inline-bookmark-link")
    except:
        li=[]
    j=0
    print(len(li))
    while(j!=len(li)):
        print(i,j)
        try:
            pp=li[j].text.split("\n")[0].split("AM ")[1]
        except:
            pp=''
        try:
            pn=li[j].text.split("\n")[1].split(": ")[1]
        except:
            pn=''
        session.append(ss)
        Rol.append(rl)
        chairs.append(ch)
        CAffs.append(ca)
        presentation.append(pp)
        presenter.append(pn)
        j+=1
    if(len(li)==0):
        pp=''
        pn=''
        session.append(ss)
        Rol.append(rl)
        chairs.append(ch)
        CAffs.append(ca)
        presentation.append(pp)
        presenter.append(pn)
        
    i+=1
```
