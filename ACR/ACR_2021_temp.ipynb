```python
def Divide(driver):
    try:
        a=[]
        i=1
        while(True):
            path="/html/body/div[2]/div/div/div/main/div/div/div[1]/p[1]/sup["+str(i)+"]"
            my_el=driver.find_element(By.XPATH,path)
            a.append(my_el.text)
            i+=1
    except:
        pass 
    
    try:
        path="/html/body/div[2]/div/div/div/main/div/div/div[1]/p[1]"
        te=driver.find_element(By.XPATH,path)
        te=te.text
    except:
        pass
    import re
    ss=re.findall(r' and |\d+|\D+',te)
    
    li=[]
    for i in ss:
        temp=i.strip().strip(",").strip()
        if(temp==' and ' or temp==',' or temp=="" or temp=='and'):
            continue
        else:
            li.append(temp)
    x=[]
    i=0
    j=0
    while(i!=len(a) and j !=len(li)):
        if(a[i]==li[j]):
            x.append(li[j])
            i+=1
            j+=1
        elif(li[j].isnumeric()):
            if(li[j][:len(a[i])]==a[i]):
                x.append(li[j][:len(a[i])])
                x.append(li[j][len(a[i]):]+li[j+1])
                j+=2
                i+=1
            else:
                if(li[j-1].isnumeric()):
                    x.append(li[j]+" "+li[j+1])
                else:
                    x[len(x)-1]+=(" "+li[j]+" "+li[j+1])
                j+=2 
        else:
            x.append(li[j])
            j+=1
    while(j!=len(li)):
        if(x[len(x)-1].isnumeric()):
            x.append(li[j])
        else:
            x[len(x)-1]+=li[j]
        j+=1
    auth=[]
    affil=[]
    c=0
    for i in range(len(x)):
        if(x[i-1].isnumeric() and x[i].isnumeric()):
            c=1
        if(c==0):
            auth.append(x[i])
        else:
            affil.append(x[i])
    M_L=[]
    k=[]
    i=0
    while(i!=len(auth)):
        l=[]
        if(auth[i].isnumeric()):
            for j in range(len(affil)):
                if(not(affil[j].isnumeric())):
                    continue
                if(auth[i]==affil[j]):
                    l.append(auth[i-1])
                    l.append(affil[j+1])
            M_L.append(l)
        i+=1
    return M_L
```

```python
Initial=[]
First=[]
Middle=[]
Last=[]
```

```python
def NameDivider(Name):
    inti=""
    first=""
    last=""
    midd=[]
    temp=Name.split(" ")
    for i in range(len(temp)):
        if(len(temp[i])==2 and i==1):
            if(temp[i][1]=='.'):
                inti=temp[i][0]
            continue
        if(i==0):
           first=temp[i]
           continue
        if(i==len(temp)-1):
           last=temp[i]
           continue
        else:
           midd.append(temp[i])
    if(midd==[]):
           mid=""
    if(midd!=[]):
           mid=" ".join(midd)
    return inti,first,mid,last


for j in range(len(names)):
    inti,first,mid,lst=NameDivider(names[j])
    Initial.append(inti)
    First.append(first)
    Middle.append(mid)
    Last.append(lst)
```
