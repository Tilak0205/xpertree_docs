```python
tt = pd.ExcelFile('ESMO_2021.xlsx').parse('Sheet1')
sub=tt['SUB_SESSION']
names=tt['NAMES']
```

```python
Initial=[]
First=[]
Middle=[]
Last=[]
```

```python
def NameDivider(Name):
    inti=""
    first=""
    last=""
    midd=[]
    temp=str(Name).split(" ")
    for i in range(len(temp)):
        if(len(temp[i])==2 and i==1):
            if(temp[i][1]=='.'):
                inti=temp[i][0]
            continue
        if(i==0):
           first=temp[i]
           continue
        if(i==len(temp)-1):
           last=temp[i]
           continue
        else:
           midd.append(temp[i])
    if(midd==[]):
           mid=""
    if(midd!=[]):
           mid=" ".join(midd)
    return inti,first,mid,last


for j in range(len(names)):
    inti,first,mid,lst=NameDivider(names[j])
    Initial.append(inti)
    First.append(first)
    Middle.append(mid)
    Last.append(lst)
```

```python
search=['Gadolinium',
'MRI+Contrast Agent',
'Focal liver lesions',
'Liver metastases',
'Nephrogenic Systemic Fibrosis',
'Colorectal Cancer+Liver Metastasis'
'Breast Cancer+Liver Metastasis',
'Lung Cancer+Liver Metastatsis',
'Gastric Cancer+Liver Metastasis',
'Primary Liver Cancer',
'Gadolinium toxicity',
'Gadolinium deposition']
```

```python
s_ele=[]
found=[]
```

```python
for i in range(len(sub)):
    f=-1
    for j in range(len(search)):
        if(('+' not in search[j]) and (type(sub[i])!=float)):
            if(search[j].lower() in sub[i].lower()):
                f=j
        else:
            if(type(sub[i])!=float):
                temp=search[j].split('+')
                if((temp[0] in sub[i].lower()) and (temp[1] in sub[i].lower())):
                    f=j
    if(f!=-1):
        s_ele.append(search[j])
        found.append('Found')
    else:
        s_ele.append('No search term')
        found.append('Not found')
```

```python
bt=pd.DataFrame()
bt['Session ST']=found
bt['InSession']=s_ele
# bt['Abstract ST']=ab_found
# bt['InAbstract']=ab_s_ele
bt['Initial']=Initial
bt['First Name']=First
bt['Middle Name']=Middle
bt['Last Name']=Last
```

```python
bt.to_excel('search_terms_2021.xlsx')
```
